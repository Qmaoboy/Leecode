# 2. Add Two Numbers
You are given two non-empty linked lists representing two non-negative integers. The digits are stored in reverse order, and each of their nodes contains a single digit. Add the two numbers and return the sum as a linked list.

You may assume the two numbers do not contain any leading zero, except the number 0 itself.  
  
Input: l1 = [2,4,3], l2 = [5,6,4]  
Output: [7,0,8]  
Explanation: 342 + 465 = 807.  
Example 2:  

Input: l1 = [0], l2 = [0]  
Output: [0]  
Example 3:  

Input: l1 = [9,9,9,9,9,9,9], l2 = [9,9,9,9]  
Output: [8,9,9,9,0,0,0,1]  

```python
from typing import Optional
# Definition for singly-linked list.
class ListNode:
    def __init__(self, val=0, next=None):
        self.val = val
        self.next = next
class Solution:
    def addTwoNumbers(self, l1: Optional[ListNode], l2: Optional[ListNode]) -> Optional[ListNode]:
        remain=0
        quotient=0
        head=ListNode(0,None)
        dummy=ListNode(0,head)
        while l1 or l2:
            linesum =int(l1.val if l1!=None else 0)+int(l2.val if l2!=None else 0)+quotient
            quotient=(linesum)//10
            remain =(linesum)%10
            if l1:
                l1=l1.next
            if l2:
                l2=l2.next
            nextlinesum =int(l1.val if l1!=None else 0)+int(l2.val if l2!=None else 0)+quotient

            if nextlinesum or l1 or l2:
                head.next=ListNode(quotient,None)
            if linesum or l1 or l2:
                head.val=remain
                head=head.next

        return dummy.next
```
